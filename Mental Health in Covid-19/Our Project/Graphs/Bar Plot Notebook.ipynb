```python
import pandas as pd 
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

CB91_Blue = '#2CBDFE'
CB91_Green = '#47DBCD'
CB91_Pink = '#F3A0F2'
CB91_Purple = '#9D2EC5'
CB91_Violet = '#661D98'
CB91_Amber = '#F5B14C'

color_list = [CB91_Blue, CB91_Pink, CB91_Green, CB91_Amber,
              CB91_Purple, CB91_Violet]
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
```

```python
data=pd.read_csv("Final Bar Graph Plot Sheet Full.csv")
```

```python
xx=5
labels = data["Category"][0:xx]
cse=data["Under"][0:xx]
ece=data["Normal"][0:xx]
cse2=data["Over"][0:xx]
ece2=data["Obese"][0:xx]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, cse, width/2, label='Underweight')
rects2 = ax.bar(x - width/2, ece, width/2, label='Normal')
rects3 = ax.bar(x , cse2, width/2, label='Overweight')
rects4 = ax.bar(x + width/2, ece2, width/2, label='Obese')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Students')
ax.set_title('Number of Students by BMI Category')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)
#ax.bar_label(rects3, padding=3)

fig.tight_layout()

plt.show()
```
